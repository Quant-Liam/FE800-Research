```python

#nelson siegel funct 

def nelson_siegel(t, beta0, beta1, beta2, tau):
    
    t = np.array(t)
    f1 = (1 - np.exp(-t / tau)) / (t / tau)
    f2 = f1 - np.exp(-t / tau)
    
    return beta0 + beta1 * f1 + beta2 * f2


#valuation date, cashflow t 

pricing_date = datetime(2020, 4, 29)

df_schedule_with_rates["t"] = (
    (df_schedule_with_rates["date"] - pricing_date).dt.days / 365)


df_schedule_with_rates = (
    
    df_schedule_with_rates[df_schedule_with_rates["t"] > 0]
    .reset_index(drop=True)
)

#nelson siegel curve 

# params for FRED curve (beta0, beta1, beta2, tau)

#idk how these are calibrated specifically 
params_ns = [0.01279104, -0.01217388, -0.00688289, 4.09851009]

#grid for smooth curve

t_grid = np.linspace(0.1, 10, 200)
y_grid = nelson_siegel(t_grid, * params_ns)

#using the tenors 3m to 10y, inputting the observed rates to work with 

tenors = np.array([0.25, 0.5, 1, 2, 3, 5, 7, 10])
rates  = np.array([0.09, 0.05, 0.17, 0.20, 0.24, 0.37, 0.51, 0.62]) / 100.0


#ns fitted yields at tenors 

ns_fitted = nelson_siegel(tenors, *params_ns)


#ns fit plot 

plt.plot(t_grid, y_grid * 100,label="Nelsonâ€“Siegel Fit", color="red", linewidth=1.8)
plt.scatter(tenors, rates * 100,label="Observed Yields", marker="o", color="blue")

plt.xlabel("Tenor (Years)")
plt.ylabel("Yield (%)")
plt.title("Treasury Yield Curve")

plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


#observed vs ns fit table 

yield_table = pd.DataFrame({
    "Tenor (Years)": tenors,
    "Observed Yield (%)": (rates * 100).round(3),
    "NS Fitted Yield (%)": (ns_fitted * 100).round(3)
})

print(yield_table.to_string(index=False))


```
